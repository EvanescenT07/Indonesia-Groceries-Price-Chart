```python
# Cleaning Pricing Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data xlsx
data = pd.read_excel('harga.xlsx')

# Check the data
data.head()
```

```python
# Check the data type
data.dtypes
```

```python
# Check missing values and drop it
data.isnull().sum()

# Drop missing values
data = data.dropna()

# Check the data
data.isnull().sum()

# Check the data
data.head()
```

```python
# Save the cleaned data to csv and json
data.to_csv('harga_cleaned.csv', index=False)
data.to_json('harga_cleaned.json', orient='records')
```
